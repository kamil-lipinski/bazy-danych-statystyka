```python
data = pd.read_csv('MDR_RR_TB_burden_estimates_2023-12-04.csv', sep=',', na_values='')
data
```

```python
print("Wartość maksymalna 'e_rr_pct_new':", data['e_rr_pct_new'].max())
print("Wartość minimalna 'e_rr_pct_new':", data['e_rr_pct_new'].min())
print("Średnia wartość 'e_rr_pct_new':", data['e_rr_pct_new'].mean())
print("Odchylenie standardowe 'e_rr_pct_new':", data['e_rr_pct_new'].std())
print("Mediana 'e_rr_pct_new':", np.median(data['e_rr_pct_new']))
print("Górna granica mediany 'e_rr_pct_new':", statistics.median_high(data['e_rr_pct_new']))
print("Dolna granica mediany 'e_rr_pct_new':", statistics.median_low(data['e_rr_pct_new']))
print("Wartości 'e_rr_pct_new' najczęściej występujące:", statistics.mode(data['e_rr_pct_new']))
print("Wariancja 'e_rr_pct_new':", statistics.variance(data['e_rr_pct_new']))
print("Odchylenie standardowe 'e_rr_pct_new':", statistics.stdev(data['e_rr_pct_new']))
print("Miara koncentracji 'e_rr_pct_new':", ss.kurtosis(data['e_rr_pct_new']))
print("Skośność rozkładu 'e_rr_pct_new':", ss.skew(data['e_rr_pct_new']))
```

2. Wgraj plik Wzrost.csv. Użyj dla niego funkcji z modułu statistics. Czym różnią się funkcje dotyczące wariancji i odchylenia standardowego? 


```python
wzrost = np.loadtxt('wzrost.csv', delimiter=',')
print(wzrost)
```

```python
print("Górna granica mediany:", statistics.median_high(wzrost))
print("Dolna granica mediany:", statistics.median_low(wzrost))
print("Wartości najczęściej występujące:", statistics.mode(wzrost))
print("Wariancja funkcja pvariance():", statistics.pvariance(wzrost))
print("Wariancja funkcja variance():", statistics.variance(wzrost))
print("Odchylenie standardowe funkcja pstdev():", statistics.pstdev(wzrost))
print("Odchylenie standardowe funkcja stdev():", statistics.stdev(wzrost))
```

* Funkcja pstdev() oblicza odchylenie standardowe populacji (population standard deviation), co oznacza, że zakłada się, iż dane reprezentują całą populację.
* Funkcja stdev() oblicza odchylenie standardowe próbki (sample standard deviation), co oznacza, że dane są traktowane jako próbka populacji.
* Funkcja pvariance() oblicza wariancje populacji (population variance), co oznacza, że zakłada się, iż dane reprezentują całą populację.
* Funkcja variance() oblicza wariancje próbki (sample variance), co oznacza, że dane są traktowane jako próbka populacji.

3. Znajdź dowolny plik csv. Zastosuj do niego statystyki z biblioteki scipy.stats. Sprawdź, jakie inne statystyki opisowe można znaleźć w tej bibliotece. 

```python
data = pd.read_csv('brain_size.csv', sep=';', na_values='.').iloc[:, 1:]
data = data.dropna()
data.head()
```

```python
print(f"Statystyka opisowa 'FSIQ': {ss.describe(data['FSIQ'])}")
print(f"Miara koncentracji: {ss.kurtosis(data.iloc[:, 1:])}")
print(f"Skośność rozkładu': {ss.skew(data.iloc[:, 1:])}")
print(f"25-ty percentyl 'FSIQ': {ss.scoreatpercentile(data['FSIQ'], 25)}")
print(f"Moda: {ss.mode(data.iloc[:, 1:])}")
```

4. Załaduj plik brain_size.csv do pandas DataFrame. Znajdź średnią dla kolumny VIQ. Ile kobiet i mężczyzn jest wyróżnionych w pliku. Wyświetl histogramy dla zmiennych VIQ, PIQ, FSIQ. Wyświetl histogramy trzech kolumn tylko dla kobiet. 

```python
data = pd.read_csv('brain_size.csv', sep=';', na_values='.').iloc[:, 1:]
data.head()
```

```python
print("Średnia wartość 'VIQ':", data['VIQ'].mean())
```

```python
gender_counts = data.Gender.value_counts()
print(f"W pliku wyróżnionych jest {gender_counts['Male']} mężczyzn i {gender_counts['Female']} kobiet")
```

```python
import matplotlib.pyplot as plt

data[['VIQ', 'PIQ', 'FSIQ']].hist(bins=20, figsize=(10, 6), grid=False)
plt.suptitle('Histogramy VIQ, PIQ, FSIQ')
plt.show()
```

```python
grouped_data = data.groupby('Gender')
female_data = grouped_data.get_group('Female')

female_data[['VIQ', 'PIQ', 'FSIQ']].hist(bins=20, figsize=(10, 6), grid=False)
plt.suptitle('Histogramy VIQ, PIQ, FSIQ kobiet')
plt.show()
```
