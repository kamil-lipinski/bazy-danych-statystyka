```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import statistics
```

```python
data = pd.read_csv('D:/155069/cw1/MDR_RR_TB_burden_estimates_2023-11-28.csv', sep=',', na_values='')
data
```

```python
print("Wartość maksymalna e_rr_pct_new:", data['e_rr_pct_new'].max())
print("Wartość minimalna e_rr_pct_new:", data['e_rr_pct_new'].min())
print("Średni wartość e_rr_pct_new:", data['e_rr_pct_new'].mean())
print("Odchylenie standardowe e_rr_pct_new:", data['e_rr_pct_new'].std())
print("Mediana e_rr_pct_new:", np.median(data['e_rr_pct_new']))
print("Górna granica mediany e_rr_pct_new:", statistics.median_high(data['e_rr_pct_new']))
print("Dolna granica mediany e_rr_pct_new:", statistics.median_low(data['e_rr_pct_new']))
print("Wartości e_rr_pct_new najczęściej występujące:", statistics.mode(data['e_rr_pct_new']))
print("Wariancja e_rr_pct_new:", statistics.variance(data['e_rr_pct_new']))
print("Odchylenie standardowe e_rr_pct_new:", statistics.stdev(data['e_rr_pct_new']))
print("Miara koncentracji e_rr_pct_new:", ss.kurtosis(data['e_rr_pct_new']))
print("Skośność rozkładu e_rr_pct_new:", ss.skew(data['e_rr_pct_new']))
```

```python
wzrost = np.loadtxt("D:/155069/cw1/wzrost.csv", delimiter=',', dtype=int)
print(wzrost)
```

```python
print("Górna granica mediany:", statistics.median_high(wzrost))
print("Dolna granica mediany:", statistics.median_low(wzrost))
print("Wartości najczęściej występujące:", statistics.mode(wzrost))
print("Wariancja funkcja pvariance():", statistics.pvariance(wzrost))
print("Wariancja funkcja variance():", statistics.variance(wzrost))
print("Odchylenie standardowe funkcja pstdev():", statistics.pstdev(wzrost))
print("Odchylenie standardowe funkcja stdev():", statistics.stdev(wzrost))
```

Funkcja pstdev() oblicza odchylenie standardowe w populacji – liczba którą można obliczyć dokładnie, jeśli znane są wartości zmiennej dla wszystkich obiektów w badanej populacji. Odpowiada ono odchyleniu zmiennej losowej, której rozkład jest taki sam jak w rozkładzie w populacji.

Funkcja stdev() oblicza odchylenie standardowe z próby – to oszacowanie odchylenia standardowego w populacji na podstawie próby losowej (znajomości zaledwie części z jej obiektów). Wzory stosowane do tego celu nazywa się estymatorami odchylenia standardowego.
