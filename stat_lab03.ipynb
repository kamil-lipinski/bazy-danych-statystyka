1. Wygeneruj próbę losową dla rozkładu normalnego dla średniej = 2, odchylenia = 30 i liczby elementów = 200. Zbadaj hipotezę mówiącą o tym, że średnia tego rozkładu jest równa 2,5. 


```python
import numpy as np
import scipy.stats as sps
```

```python
srednia = 2
odchylenie = 30
n = 200

data = np.random.normal(srednia, odchylenie, n)

hipoteza = 2.5
stat, p_value = sps.ttest_1samp(data, hipoteza)

print(f'Statystyka testowa: {stat}')
print(f'Wartość p: {p_value}')

if p_value < 0.05:
    print("Odrzucamy hipotezę zerową.")
else:
    print("Nie ma podstaw do odrzucenia hipotezy zerowej.")
```

2. Wczytaj plik napoje.csv do pandas DataFrame. Zweryfikuj hipotezę że średnie spożycie piwa lech wynosi 60500, coli wynosi 222000, piw regionalne wynosi 43500.

```python
import pandas as pd
```

```python
df = pd.read_csv('napoje.csv', sep = ';')
df
```

```python
hipotezy = {'lech': 60500, 'cola': 222000, 'regionalne': 43500}

for piwo, hipoteza in hipotezy.items():
    data = df[piwo]
    stat, p_value = sps.ttest_1samp(data, hipoteza)
    
    print('\n-----------------------------')
    print(f'Test dla: {piwo}')
    print(f'Statystyka testowa: {stat}')
    print(f'Wartość p: {p_value}')

    if p_value < 0.05:
        print(f'Odrzucamy hipotezę zerową.')
    else:
        print('Nie ma podstaw do odrzucenia hipotezy zerowej.')
```

3. Sprawdzić która zmienna w pliku napoje.csv wykazuje normalność

```python
for zmienna in df.columns:  
    
    data = df[zmienna]
    stat, p_value = sps.shapiro(data)

    print('\n-----------------------------')
    print(f'Test normalności dla zmiennej {zmienna}:')
    print(f'Statystyka testowa: {stat}')
    print(f'Wartość p: {p_value}')

    if p_value < 0.05:
        print(f'Rozkład dla {zmienna} nie wykazuje normalności')
    else:
        print(f'Rozkład dla {zmienna} wykazuje normalność')
```

4. Zbadaj równość średnich dla następujących par: okocim – lech, fanta – regionalne oraz cola – pepsi. 

```python
pary = [('okocim', 'lech'), ('fanta ', 'regionalne'), ('cola', 'pepsi')]

for zmienna1, zmienna2 in pary:
    
    data1 = df[zmienna1]
    data2 = df[zmienna2]
    
    stat, p_value = sps.mannwhitneyu(data1, data2)
    
    print('\n-----------------------------')
    print(f'Test równości średniej dla par {zmienna1} - {zmienna2}:')
    print(f'Statystyka testowa: {stat}')
    print(f'Wartość p: {p_value}')

    if p_value < 0.05:
        print(f'Średnie {zmienna1} i {zmienna2} nie są równe.')
    else:
        print(f'Średnie {zmienna1} i {zmienna2} są równe.')
```

5. Zbadaj równość wariancji pomiędzy okocim – lech, żywiec – fanta oraz regionalne – cola. 

```python
pary = [('okocim', 'lech'), ('żywiec', 'fanta '), ('regionalne', 'cola')]

for zmienna1, zmienna2 in pary:
    
    data1 = df[zmienna1]
    data2 = df[zmienna2]
    
    stat, p_value = sps.levene(data1, data2)
    
    print('\n-----------------------------')
    print(f'Test równości wariancji dla par {zmienna1} - {zmienna2}:')
    print(f'Statystyka testowa: {stat}')
    print(f'Wartość p: {p_value}')

    if p_value < 0.05:
        print(f'Wariancje {zmienna1} i {zmienna2} nie są równe.')
    else:
        print(f'Wariancje {zmienna1} i {zmienna2} są równe.')
```

6. Zbadaj równość średnich pomiędzy latami 2001 i 2015 dla piw regionalnych. 

```python
grouped_data = df.groupby('rok')

regionalne_2001 = grouped_data.get_group(2001)['regionalne']
regionalne_2015 = grouped_data.get_group(2015)['regionalne']

stat, p_value = sps.mannwhitneyu(regionalne_2001, regionalne_2015)
    
print(f'Test równości średniej pomiedzy latami 2001 i 2015 dla piw regionalnych:')
print(f'Statystyka testowa: {stat}')
print(f'Wartość p: {p_value}')

if p_value < 0.05:
    print(f'Średnie pomiedzy latami 2001 i 2015 nie są równe.')
else:
    print(f'Średnie pomiedzy latami 2001 i 2015 są równe.')
    
```

7. Zbadaj równość średnich dla wartości z roku 2016 oraz dla wartości z pliku napoje_po_reklamie.csv oddzielnie dla coli, fanty i pepsi. Zakładamy, że zmienne te są zależne.

```python
df_po_reklamie = pd.read_csv('napoje_po_reklamie.csv', sep = ';')
df_po_reklamie
```

```python
for zmienna in ['cola', 'fanta ', 'pepsi']:
    
    data_2016 = grouped_data.get_group(2016)[zmienna]
    data_po_reklamie = df_po_reklamie[zmienna]
    
    stat, p_value = sps.mannwhitneyu(data_2016, data_po_reklamie)
    
    print('\n-----------------------------')
    print(f'Test równości średniej dla {zmienna} z roku 2016 i po reklamie:')
    print(f'Statystyka testowa: {stat}')
    print(f'Wartość p: {p_value}')

    if p_value < 0.05:
        print(f'Średnie {zmienna} z roku 2016 i po reklamie nie są równe.')
    else:
        print(f'Średnie {zmienna} z roku 2016 i po reklamie są równe.')
```
