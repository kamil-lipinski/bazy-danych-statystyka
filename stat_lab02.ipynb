1. Dla zmiennych losowych przedstawionych w tabeli obliczyć podstawowe statystyki.

Wartość | 1 | 2 | 3 | 4 | 5 | 6 |

Prawdopodobieństwo | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 |



```python
import numpy as np
```

```python
wartosci = np.array([1, 2, 3, 4, 5, 6])
prawdopodobienstwo = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

srednia = np.sum(wartosci * prawdopodobienstwo)
print(f"Średnia: {srednia}")

wariancja = np.sum((wartosci - srednia) ** 2 * prawdopodobienstwo)
print(f"Wariancja: {wariancja}")

odchylenie = np.sqrt(wariancja)
print(f"Odchylenie standardowe: {odchylenie}")
```

2. Wygeneruj próby dla n=100 dla następujących rozkładów: Bernoulliego, Dwumianowego, Poissona
3. Policz dla nich statystyki podstawowe (średnią, wariancję, kurtozę i skośność).


```python
import scipy.stats as scs
```

```python
# Rozkład Bernoulliego

p_bernoulli = 0.3
bernoulli_samples = scs.bernoulli.rvs(p_bernoulli, size=100)
mean_bernoulli, var_bernoulli, skew_bernoulli, kurt_bernoulli = scs.bernoulli.stats(
    p_bernoulli, moments='mvsk')

print(f"Średnia: {mean_bernoulli}")
print(f"Wariancja: {var_bernoulli}")
print(f"Skosność: {skew_bernoulli}")
print(f"Kurtoza: {kurt_bernoulli}")
```

```python
# Rozkład Dwumianowy

n_binomial = 10
p_binomial = 0.3
binomial_samples = scs.binom.rvs(n_binomial, p_binomial, size=100)
mean_binomial, var_binomial, skew_binomial, kurt_binomial = scs.binom.stats(
    n_binomial, p_binomial, moments='mvsk')

print(f"Średnia: {mean_binomial}")
print(f"Wariancja: {var_binomial}")
print(f"Skosność: {skew_binomial}")
print(f"Kurtoza: {kurt_binomial}")
```

```python
# Rozkład Poissona

lambda_poisson = 3
poisson_samples = scs.poisson.rvs(lambda_poisson, size=100)
mean_poisson, var_poisson, skew_poisson, kurt_poisson = scs.poisson.stats(lambda_poisson, moments='mvsk')

print(f"Średnia: {mean_poisson}")
print(f"Wariancja: {var_poisson}")
print(f"Skosność: {skew_poisson}")
print(f"Kurtoza: {kurt_poisson}")
```

4. Dla rozkładów Bernoulliego, Dwumianowego i Poissona narysuj wykres składający się z 3 części, gdzie każda część będzie zawierać wykres rozkładu prawdopodobieństwa 

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 5))

# Rozkład Bernoulliego
plt.subplot(1, 3, 1)
x_bernoulli = np.array([0, 1])
bernoulli_pmf = scs.bernoulli.pmf(x_bernoulli, p_bernoulli)
plt.bar(x_bernoulli, bernoulli_pmf)
plt.title('Rozkład Bernoulliego')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')

# Rozkład Dwumianowy
plt.subplot(1, 3, 2)
x_binomial = np.arange(0, n_binomial + 1)
binomial_pmf = scs.binom.pmf(x_binomial, n_binomial, p_binomial)
plt.bar(x_binomial, binomial_pmf)
plt.title('Rozkład Dwumianowy')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')

# Rozkład Poissona
plt.subplot(1, 3, 3)
x_poisson = np.arange(0, 15)
poisson_pmf = scs.poisson.pmf(x_poisson, lambda_poisson)
plt.bar(x_poisson, poisson_pmf)
plt.title('Rozkład Poissona')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')

plt.show()
```

5. Dla rozkładu Dwumianowego wygeneruj rozkład prawdopodobieństwa dla n = 20, k = 0, …, 20 oraz p = 0.4. Sprawdź, czy suma prawdopodobieństw wygenerowana dla rozkładu dwumianowego jest równa 1. 


```python
n_binomial = 20
p_binomial = 0.4

k_binomial = np.arange(0, 21)
binomial_pmf = scs.binom.pmf(k_binomial, n_binomial, p_binomial)

plt.bar(k_binomial, binomial_pmf)
plt.title('Rozkład Dwumianowy')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')
plt.show()


sum_probability = np.sum(binomial_pmf)
print(f"Suma prawdopodobieństw: {sum_probability:.2f}")
```

Suma równa 1

6. Wygeneruj 100 danych dla rozkładu normalnego o średniej równej 0 i odchyleniu standardowym równym 2. Wyznacz wszystkie statystyki podstawowe – czy są one równe z wartościami teoretycznymi? Sprawdź, czy zwiększenie liczby danych zwiększy dokładność wyliczeń statystyk opisowych. 

```python
srednia = 0
odchylenie = 2
n = 100

data = np.random.normal(srednia, odchylenie, n)

mean_data = np.mean(data)
var_data = np.var(data)
std_data = np.std(data)
skew_data = scs.skew(data)
kurt_data = scs.kurtosis(data)

theoretical_mean = srednia
theoretical_std = odchylenie
theoretical_var = odchylenie**2
theoretical_skew = 0
theoretical_kurt = 0

print("Statystyki podstawowe:")
print(f"Średnia danych: {mean_data}, Teoretyczna średnia: {theoretical_mean}")
print(f"Wariancja danych: {var_data}, Teoretyczna wariancja: {theoretical_var}")
print(f"Odchylenie standardowe danych: {std_data}, Teoretyczne odchylenie standardowe: {theoretical_std}")
print(f"Skośność danych: {skew_data}, Teoretyczna skośność: {theoretical_skew}")
print(f"Kurtoza danych: {kurt_data}, Teoretyczna kurtoza: {theoretical_kurt}")

n2 = 10000
new_data = np.random.normal(srednia, odchylenie, n2)

new_mean_data = np.mean(new_data)
new_var_data = np.var(new_data)
new_std_data = np.std(new_data)
new_skew_data = scs.skew(new_data)
new_kurt_data = scs.kurtosis(new_data)

print("\nStatystyki podstawowe n=10000:")
print(f"Średnia danych: {new_mean_data}")
print(f"Wariancja danych: {new_var_data}")
print(f"Odchylenie standardowe danych: {new_std_data}")
print(f"Skośność danych: {new_skew_data}")
print(f"Kurtoza danych: {new_kurt_data}")
```

Zwiększenie liczby danych zwiększa dokładność wyliczeń statystyk opisowych, statystyki dla większej probki sa znacznie bliższe statystykom teoretycznym.

7. Narysować na jednym wykresie histogram dla rozkładu normalnego o parametrach: średnia = 1, odchylenie =2, wykres dla rozkładu standardowego, oraz wykres gęstości dla średniej równej -1 oraz odchylenia równego 0.5

```python
mean_normal1 = 1
std_normal1 = 2

mean_standard = 0
std_standard = 1

mean_normal2 = -1
std_normal2 = 0.5

data_normal1 = np.random.normal(mean_normal1, std_normal1, 1000)
data_standard = np.random.normal(mean_standard, std_standard, 1000)
data_normal2 = np.random.normal(mean_normal2, std_normal2, 1000)

plt.hist(data_normal1, bins=30, density=True, alpha=0.5, color='blue', label='Normalny(1, 2)')

x = np.linspace(-8, 8, 100)
pdf_standard = scs.norm.pdf(x, mean_standard, std_standard)
pdf_normal2 = scs.norm.pdf(x, mean_normal2, std_normal2)
plt.plot(x, pdf_standard, 'green', linewidth=2,  label='Normalny standardowy')
plt.plot(x, pdf_normal2, 'red', linewidth=2, label='Normalny(-1, 0.5)')

plt.xlabel('Wartość')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.legend()
plt.show()
```
